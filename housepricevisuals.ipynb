```python
import pandas as pd

# Load the data
df = pd.read_csv('karachi_house.csv', encoding='ascii')

# Conversion factors
to_square_yard = {'Kanal': 605, 'Marla': 30.25}

# Function to convert area to square yards
def convert_to_square_yard(row):
    area_size = row['Area Size']
    area_type = row['Area Type']
    return area_size * to_square_yard.get(area_type, 1)

# Apply the conversion
df['area_square_yard'] = df.apply(convert_to_square_yard, axis=1)

# Display the updated dataframe head
print(df.head())
```

```python
dropp=['Area Size','Area Type','property_type','city','Area Category']
```

```python
df=df.drop(dropp,axis=1)
```

```python
df.head()
```

```python
df=df.drop('area',axis=1)
```

```python

df = df.rename(columns={'area_square_yard': 'AreaSqYards'})
df.head()
```

```python

import pandas as pd
# Group data by location and calculate the average price for each location in 2019 and 2023
average_price_2019 = df2019.groupby('location')['price'].mean()
average_price_2023 = df2023.groupby('location')['price'].mean()

# Combine the average prices into a single DataFrame
price_comparison = pd.DataFrame({'Average Price 2019': average_price_2019, 'Average Price 2023': average_price_2023})

# Calculate the percentage change in price
price_comparison['Percentage Change'] = ((price_comparison['Average Price 2023'] - price_comparison['Average Price 2019']) / price_comparison['Average Price 2019']) * 100

# Display the price comparison
print(price_comparison)
```

```python

def categorize_area(area):
  if 60 <= area <= 240:
    return '60-240'
  elif 240 < area <= 500:
    return '240-500'
  elif 500 < area <= 1000:
    return '500-1000'
  elif 1000 < area <= 1500:
    return '1000-1500'
  elif 1500 < area <= 2000:
    return '1500-2000'
  else:
    return '2000+'

df2023['AreaSqYardsCategory'] = df2023['AreaSqYards'].apply(categorize_area)
df2023.head()
```

```python
df2019['AreaSqYardsCategory'] = df2019['AreaSqYards'].apply(categorize_area)
df2019.head()
```

```python

df2019['AreaSqYardsCategory'].unique()
df2023['AreaSqYardsCategory'].unique()
```

```python
# Merge the datasets on 'location' and 'area_category'
merged_df = pd.merge(df2019, df2023, on=['location', 'AreaSqYardsCategory'], suffixes=('_2019', '_2023'))

# Check the merged data
merged_df.head()
```

```python
# Group by location and area category to calculate the average price for 2019 and 2023
grouped_df = merged_df.groupby(['location', 'AreaSqYardsCategory']).agg(
    avg_price_2019=('price_2019', 'mean'),
    avg_price_2023=('price_2023', 'mean')
).reset_index()

# Calculate the percentage change in price
grouped_df['percentage_change'] = ((grouped_df['avg_price_2023'] - grouped_df['avg_price_2019']) / grouped_df['avg_price_2019']) * 100

# Check the final result
grouped_df.head()
```

```python
grouped_df.to_csv('comparision.csv', index=False)
```

```python

Q1 = grouped_df['percentage_change'].quantile(0.25)
Q3 = grouped_df['percentage_change'].quantile(0.75)
IQR = Q3 - Q1

# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove rows with outliers in 'percentage_change'
grouped_df_no_outliers = grouped_df[
    (grouped_df['percentage_change'] >= lower_bound) &
    (grouped_df['percentage_change'] <= upper_bound)
]

# Print the DataFrame without outliers
print(grouped_df_no_outliers)
```
